```python
%autoreload 2
import rnn as rn

import os
from email import parser as ep

parser = ep.Parser()

data = ''
cnt = 0
skip = 2001
limit = 2
maildir = 'maildir/'
for root, dirs, files in os.walk(maildir):
    if cnt > limit+skip:
        break
    stripped = root[len(maildir):]
    print("entering %s containing %d files and %d dirs" % (stripped, len(files), len(dirs)))

    for file in files:
        with open(root + "/" + file, "r", encoding='utf-8', errors='ignore') as f:  
            cnt += 1
            body = parser.parsestr(f.read()).get_payload()
            if skip > cnt:
                continue
            data += body + '\n'

chars = list(set(data))
data_size, vocab_size = len(data), len(chars)
print('data has %d characters, %d unique.' % (data_size, vocab_size))
print('===== DATA =====')
print(data)
print('===== END DATA =====')

char_to_ix = {ch: i for i, ch in enumerate(chars)}
ix_to_char = {i: ch for i, ch in enumerate(chars)}
```

```python
data_i = [char_to_ix[ch] for ch in data]

rnn = rn.RNN(vocab_size, hidden_size=100, sequence_length=25)
rnn.fit(data_i, ix_to_char, learning_rate=0.1, max_iter=30000, verbose=300)
```
